```python
def inverse(p):
    p = np.asarray(p, dtype = np.int32)
    assert p.ndim==1
    s = np.empty(p.size, dtype = np.int32)
    s[p] = np.arange(p.size, dtype=np.int32)
    return s
    
def fix_image(rbga, degree):
    # assumes its a pres ftm
    rbga = rbga.to_colexical()
    rbga.prune_above_dimension(degree+1)
    B = rbga.get_boundaries()
    F = rbga.get_filtrations()
    dims = rbga.get_dimensions()
    assert dims[0] == degree and dims[-1] == degree+1
    
    num_gen = np.sum(dims==degree)
    num_rel = np.sum(dims==degree+1)
    B_ =  list([[i] for i in range(num_gen)]) + list(B[num_gen:])
    dims_ = [degree]*num_gen + [degree]*(num_rel) # put everything to the same dim such that they can interact
    F_ = np.concatenate([F[:num_gen], F[num_gen:]])
    im_ = mp.Slicer(rbga, return_type_only=True, )(B_, dims_, F_)
    im_,p = im_.to_colexical(return_permutation =True)
    new_s = im_.compute_kernel_projective_cover(degree-1) # beta

    p = np.concatenate([inverse(p)[:num_gen], np.arange(num_gen+num_rel, len(new_s), dtype=np.int32)])
    new_s = new_s.permute_generators(p) # pi_0
    return new_s
    
def _rb_ga(a,b,dim=1):
    idx = np.argwhere(a.get_dimensions() == dim+1)
    Fa = np.asarray(a.get_filtrations())
    Fb = np.asarray(b.get_filtrations())
    B = a.get_boundaries()
    dims = a.get_dimensions()
    F = np.array(Fa)
    F[idx] = Fb[idx]
    c = type(a)(B,dims,F)
    return c

def to_rbga(s1, degree):
    _b = s1.get_boundaries()
    _d = s1.get_dimensions()
    _f = np.asarray(s1.get_filtrations())
    _f[:,0] /= 2
    s2 = type(s1)(_b,_d,_f)
    return _rb_ga(s1,s2,degree).to_colexical()
```

```python
def get_immuno(i=1, DATASET_PATH = expanduser("~/Datasets/")):
    immu_dataset = read_csv(DATASET_PATH+f"LargeHypoxicRegion{i}.csv")
    X = np.array(immu_dataset['x'])
    X /= np.max(X)
    Y = np.array(immu_dataset['y'])
    Y /= np.max(Y)
    # labels = LabelEncoder().fit_transform(immu_dataset['Celltype'])
    labels = immu_dataset['Celltype']
    return np.asarray([X,Y]).T, labels
```

```python
def get_s(k, i, degree=1, threshold=None, filtration='d'):
    # x = np.random.uniform(low=[0,0.], high=[1.,1], size=(k,2))
    import fpsample
    y = np.random.uniform(low=[0,0.], high=[1.,1],size=(10*k,2))
    idx = fpsample.bucket_fps_kdline_sampling(y, k, 5)
    x = np.array(y[idx])
    del y, idx
    f = -KDE(bandwidth=bandwidth, return_log=True).fit(X[Y==i]).score_samples(x)
    if filtration == 'd':
        s = mp.filtrations.DelaunayLowerstar(points=x, function=f, reduce_degree=degree)
        s = mp.Slicer(s, backend="gudhicohomology")
        # print(len(s))
    elif filtration == "ri":
        st = mp.filtrations.RipsLowerstar(points = x, function=f, threshold_radius=threshold)
        # print("Num simplices.",len(st))
        st.collapse_edges(-1)
        st.expansion(degree+1)
        s = mp.Slicer(st).minpres(degree)
        s = to_rbga(s, degree=degree)
        s = fix_image(s, degree=degree).minpres(degree)
    return s
```

```python
from itertools import product
def get_error_i(i,n):
    s = get_s(n, i,filtration='d', threshold=delta(n))
    bc = s.persistence_on_lines(basepoints=[[delta(n),0]], directions=[[0,1]])[0][degree]
    del s
    error =  gd.bottleneck_distance(
        ground_truth[i].clip(None, threshold_height),
        bc.clip(None, threshold_height)
    )
    del bc
    return error
    
num_ptss = np.logspace(3,5, 200, dtype=int)
# num_ptss = np.linspace(1e1,1e4, 200, dtype=int)
errors = Parallel(n_jobs=-1, backend="loky")(delayed(get_error_i)(
    i,n
) for i,n in tqdm(product(range(3), num_ptss), total=3*len(num_ptss)))
```

```python
from multipers.distances import sm_distance
from joblib import Parallel, delayed
from tqdm import tqdm
from multipers.vector_interface import aida
i = 2
degree = 1
invariant = "rank"
res = 500
_X = X[Y==i]
bandwidth=.05
ns = np.linspace(100, len(_X), 200, dtype=int)
densities = KDE(bandwidth=bandwidth).fit(_X).score_samples(_X)

def _todo(n, X, F):
    x = X[:n]
    codens = F[:n]
    s1 = mp.filtrations.DelaunayLowerstar(x,codens, reduce_degree=degree, vineyard=False).minpres(degree, force=True)
    s_dec = aida(s1)
    max_res=200
    sms = Parallel(n_jobs=-1, backend="threading")(
        delayed(
            lambda x: mp.signed_measure(
                (x if len(x)<max_res else x.grid_squeeze(strategy="regular_closest", resolution=max_res)),
                degree=degree,
                invariant="rectangle",
            )[0]
        )(x)
        for x in s_dec
    )
    sm = mp.point_measure.add_sms(sms)
    return sm

sms_delaunay = Parallel(n_jobs=5, backend="loky")(delayed(_todo)(n, _X, -densities) for n in tqdm(ns))

last = sms_delaunay[-1]
```

```python
for j, (a,b) in enumerate(sms_delaunay):
    if len(a) == 0:
        a = np.empty((0,4))
    a[a==np.inf] = 0
    sms_delaunay[j] = (a,b)
```

```python
from multipers.distances import sm_distance, sm2diff
from ot import emd2
import torch
def sm_distance(sm1,sm2):
    x,y = sm2diff(sm1,sm2)
    x = torch.from_numpy(x).cuda()
    y = torch.from_numpy(y).cuda()
    c = torch.cdist(x,y).cpu().numpy()
    # print(c)
    return emd2([],[],c)
errors =Parallel(backend="threading", n_jobs=10)(delayed(sm_distance)(sm,last) for sm in tqdm(sms_delaunay))
```

```python
from sklearn.linear_model import LinearRegression
stuff = LinearRegression().fit(np.asarray(ns)[:,None], np.asarray(errors))
a,(b,) = stuff.intercept_, stuff.coef_
print(a,b)
plt.plot([1, ns[-1]], [a, a+b*ns[-1]], label=fr"y=${a:.2f}{b:.2g}x$", c="k", linestyle="--")
plt.plot((ns)[:-5],(errors)[:-5])
# plt.hlines(0, 0, ns[-1], ls="--",color="k")
plt.legend()
plt.xlabel("Number of points")
plt.ylabel("Wasserstein-2 error to last")
plt.savefig(f"errors/{invariant}_error{i}_res{res}_degree{degree}_bw{bandwidth}.pdf")
```

## Matching distance error

```python
# Ground truth
res = 500
bandwidth = .05
grid = mp.grids.compute_grid([[0,1]]*2, strategy="regular", resolution=res)
grid = mp.grids.todense(grid)
def get_f(x):
    return np.array(np.vstack([-KDE(bandwidth=bandwidth, return_log=True).fit(X[Y==i]).score_samples(x) for i in range(3)]).T)
img3d =  get_f(grid).reshape(res,res,3)
D = 10*img3d.max()
ground_truth = mp.filtrations.Cubical(img3d)
ground_truth.prune_above_dimension(1)
ground_truth = mp.Slicer(ground_truth, backend="gudhicohomology")

# ac = gd.AlphaComplex(points=grid)
# st = ac.create_simplex_tree(max_alpha_square=1) 
# st.prune_above_dimension(1)
# st = mp.SimplexTreeMulti(st, num_parameters=3)
# st.lower_star_multi_filtration_update(get_f(grid))
# assert st.num_vertices == len(grid)
# ground_truth = st

ground_truth = mp.Slicer(ground_truth)
len(ground_truth)
```

```python
lower_bound, upper_bound = img3d.min(), img3d.max()
filtration_range = upper_bound-lower_bound
lower_bound, upper_bound, filtration_range
```

```python
import gc
# random lines in R^3.
nlines = 1000
random_directions=True
basepoints = lower_bound + filtration_range*np.random.uniform(size=(nlines,3))
if random_directions:
    directions = np.empty((0,3))
    while len(directions)<nlines:
        thetas = np.random.normal(size=(nlines,3))
        thetas = thetas[np.all(thetas>0, axis=1)]
        thetas /= np.min(thetas, axis=1)[:,None]
        directions = np.concatenate([directions, thetas])
    directions = directions[:nlines]
else:
    directions = np.ones(shape=(nlines,3))

degree=0
ground_truth_bcs = tuple(b[degree] for b in ground_truth.persistence_on_lines(basepoints=basepoints, directions=directions))

del ground_truth, img3d
gc.collect()
```

```python
# import numpy as np
# import matplotlib.pyplot as plt
# import fpsample
# N=10000
# x = np.random.uniform(size=(100*N,2))
# i = fpsample.bucket_fps_kdline_sampling(x, N, 10)
# x = x[i]
# plt.scatter(*x.T, s=1)
```

```python
import gudhi as gd
from joblib import Parallel, delayed
import gc

# estimator
delta = lambda n : np.sqrt(np.log10(n)/n)

ns = np.logspace(2., 5, 200, dtype=int)
errorss = []
for N in tqdm(ns):
    gc.collect()
    threshold=delta(N)
    # print(threshold)
    
    import fpsample
    y = np.random.uniform(size=(10*N,2))
    idx = fpsample.bucket_fps_kdline_sampling(y, N, 7)
    x = np.array(y[idx])
    del y
    
    
    
    s = mp.filtrations.RipsLowerstar(points=x, function=get_f(x), threshold_radius=threshold)
    basepoints_ =  np.concat([np.repeat(.999*threshold, basepoints.shape[0])[:,None], basepoints], axis=1)
    directions_ =  np.concat([np.repeat(0, directions.shape[0])[:,None], directions], axis=1)
    est_bcs = tuple(b[degree] for b in mp.Slicer(s,backend="gudhicohomology").persistence_on_lines(basepoints=basepoints_, directions=directions_))
    
    del s, basepoints_, directions_
    bc_ = np.concatenate([b for b in est_bcs])
    
    assert (bc_==np.inf).sum() == nlines, (bc_==np.inf).sum()
    del bc_
    # print(threshold, bc_)
    
    
    errors = Parallel(n_jobs = -1, backend="threading")(
        delayed(
            lambda a,b : gd.bottleneck_distance(
                a,
                b
            )
        )(a,b)
        for a,b in zip(ground_truth_bcs, est_bcs)
    )
    
    del est_bcs
    errorss.extend([errors])
    

stuff = np.asarray(errorss)
```

```python
from scipy.stats import linregress

quantile_plot = False
if quantile_plot:
    ts = np.linspace(0,1,5)
    qs = np.quantile(stuff, q=ts, axis=1)
plt.plot(ns, delta(ns), c="k", label=r"$\delta_k$", ls="--")
# plt.hlines(D, ns[0], ns[-1], color="k", ls="--", label="threshold")
if quantile_plot:
    for machin,t in zip(qs,ts):
        plt.plot(ns, machin, label = f"Quantile {t:2g}")
else:
    # for machin in zip(stuff.T):
    #     plt.plot(ns, machin[0], color="k", alpha=.01)
    # plt.scatter([],[], color="k",alpha=.1, label="line errors")
    slope, intercept, r_value, p_value, std_err = linregress(np.log10(ns), np.log10(stuff.max(1)))
    fit = 10**(np.log10(ns)*slope + intercept)
    slope = np.round(slope,decimals=2)
    intercept = np.round(intercept,decimals=2)
    plt.plot(ns, fit, c="darkblue", label=rf"Fit: $y=10^{{{intercept}}}x^{{{slope}}}$")
    plt.scatter(ns, stuff.max(1), c="darkred", label="(Estimated) matching distance")
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Sample size $k$")
plt.ylabel("Matching distance")
plt.legend()
plt.savefig(f"output/vertical_interleaving_matching_bdw{bandwidth}_error_n{ns[-1]}_res{res}_diag{not random_directions}_nlines{nlines}_quantile{quantile_plot}.pdf")
```

```python
from scipy.spatial.distance import cdist
from multipers.distances import sm_distance
from joblib import Parallel, delayed
from tqdm import tqdm
degree=0
invariant="hilbert"
res=100
_X = X[Y==i]
sample = np.random.uniform(size=(1000, 2))
D = cdist(sample,sample)
ns = np.unique(np.linspace(100, len(sample), 200, dtype=int))
densities = np.concatenate([KDE(bandwidth=bandwidth).fit(X[Y==i]).score_samples(sample)[:,None] for i in range(3)], axis=1)
sms_delaunay = []
sm=0
threshold=.1
strategy="regular_closest"
def _todo(n):
    # assert n <=200
    st = mp.filtrations.RipsLowerstar(distance_matrix=np.array(D[:n,:n]), function=-np.array(densities[:n,:]), threshold_radius=threshold)
    # assert st.num_parameters == 4, "invalid num_parameters"
    st.expansion(degree+1)
    if degree >0:
        s1 = mp.Slicer(st)
        rbga = to_rbga(s1).grid_squeeze(strategy=strategy, resolution=res)
    else:
        rbga = st.grid_squeeze(strategy=strategy, resolution=res)
    # im = fix_image(rbga, degree) # only for 2 param
    sm, = mp.signed_measure(rbga, invariant=invariant, degree=degree, ignore_infinite_filtration_values=True)
    return sm

sms_delaunay = Parallel(n_jobs=10)(delayed(_todo)(n) for n in tqdm(ns))

last = sms_delaunay[-1]
```
