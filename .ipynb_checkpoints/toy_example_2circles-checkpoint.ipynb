# Random example

```python
def sample_points_on_circles(centers, radii, n_points_per_circle, noise_scale=0.0, seed=None):
    """
    Uniformly sample points on circles with optional Gaussian noise.

    Args:
        centers (list of tuple): List of (x, y) coordinates for circle centers.
        radii (list or float): List of radii, or a single radius for all circles.
        n_points_per_circle (list of int): list of Number of points to sample per circle.
        noise_scale (float): Standard deviation of Gaussian noise added to each point.
        seed (int, optional): Random seed for reproducibility.

    Returns:
        np.ndarray: Array of shape (N, 2) of sampled points, where N = len(centers) * n_points_per_circle.
    """
    if seed is not None:
        np.random.seed(seed)
        
    if isinstance(radii, (int, float)):
        radii = [radii] * len(centers)

    all_points = []

    for (cx, cy), r, npts in zip(centers, radii, n_points_per_circle):
        # Uniformly sample angles in [0, 2Ï€)
        angles = np.random.uniform(0, 2 * np.pi, npts)
        x = cx + r * np.cos(angles)
        y = cy + r * np.sin(angles)

        points = np.stack((x, y), axis=1)

        # Add Gaussian noise if specified
        if noise_scale > 0:
            points += np.random.normal(scale=noise_scale, size=points.shape)

        all_points.append(points)

    return np.vstack(all_points)




def draw_circles_with_colored_samples(centers, radii, samples, gt, noise, cmap='viridis'):
    """
    Draw red circle outlines and color points according to y-values.
    """
    fig, ax = plt.subplots(figsize=(6, 4))
    
    if isinstance(radii, (int, float)):
        radii = [radii] * len(centers)

    # --- Draw red circle outlines ---
    for (cx, cy), r in zip(centers, radii):
        circle = plt.Circle((cx, cy), r, color='red', fill=False, lw=2)
        ax.add_patch(circle)

    # --- Scatter points colored by y-coordinate ---
    y_vals = samples[:, 1]
    sc = ax.scatter(samples[:, 0], samples[:, 1],
                    c=y_vals, cmap=cmap, s=10, alpha=0.9, edgecolors='none')

    ax.set_aspect('equal', adjustable='datalim')

    plt.hlines(ground_truth[:,0], xmin=-2,xmax=5.3, linestyles="--", color="k", lw=2)

    filename = f"./output/2circles/2circles_{'' if noise else 'no'}noise_dataset.pdf"
    plt.savefig(filename, bbox_inches='tight')
    print(f"Saved figure as {filename}")
    plt.show()
    
```

```python
noise = False
centers = [(-0.8, 0), (3, 0)]
radii = [1, 2]
npts_per_circle = [60, 140]

height_threshold = 10
ground_truth = np.array([[1,height_threshold], [2,height_threshold]])


if noise==False:
    x = sample_points_on_circles(centers, radii, npts_per_circle, noise_scale=0, seed=27)
else:
    x = sample_points_on_circles(centers, radii, npts_per_circle, noise_scale=0.1, seed=42)

draw_circles_with_colored_samples(centers, radii, x, ground_truth, noise)
```

```python
def inverse(p):
    p = np.asarray(p, dtype = np.int32)
    assert p.ndim==1
    s = np.empty(p.size, dtype = np.int32)
    s[p] = np.arange(p.size, dtype=np.int32)
    return s
    
def fix_image(rbga, degree):
    """
    Compute the presentation of the image of a morphism between finitely presentable modules.
    """
    # assumes its a pres ftm
    rbga = rbga.to_colexical()
    rbga.prune_above_dimension(degree+1)
    B = rbga.get_boundaries()
    F = rbga.get_filtrations()
    dims = rbga.get_dimensions()
    assert dims[0] == degree and dims[-1] == degree+1
    
    num_gen = np.sum(dims==degree)
    num_rel = np.sum(dims==degree+1)
    B_ =  list([[i] for i in range(num_gen)]) + list(B[num_gen:])
    dims_ = [degree]*num_gen + [degree]*(num_rel) # put everything to the same dim such that they can interact
    F_ = np.concatenate([F[:num_gen], F[num_gen:]])
    im_ = mp.Slicer(rbga, return_type_only=True, )(B_, dims_, F_)
    im_,p = im_.to_colexical(return_permutation =True)
    new_s = im_.compute_kernel_projective_cover(degree-1) # beta

    p = np.concat([inverse(p)[:num_gen], np.arange(num_gen+num_rel, len(new_s), dtype=np.int32)])
    new_s = new_s.permute_generators(p) # pi_0
    return new_s
    
def _rb_ga(a,b,dim=1):
    idx = np.argwhere(a.get_dimensions() == dim+1)
    Fa = np.asarray(a.get_filtrations())
    Fb = np.asarray(b.get_filtrations())
    B = a.get_boundaries()
    dims = a.get_dimensions()
    F = np.array(Fa)
    F[idx] = Fb[idx]
    c = type(a)(B,dims,F)
    return c

def to_rbga(s1, degree):
    _b = s1.get_boundaries()
    _d = s1.get_dimensions()
    _f = np.asarray(s1.get_filtrations())
    _f[:,0] /= 2
    s2 = type(s1)(_b,_d,_f)
    return _rb_ga(s1,s2,degree).to_colexical()
```

```python
threshold_radius = 2.5
box = [[0,-2],[threshold_radius,5]]
codensity = x[:,1]
degree = 1

basepoints = np.linspace(start=[0., 0], stop=[threshold_radius, 0], num=500)
directions = np.asarray([[0,1]]*len(basepoints))
```

## $H_1(\mathcal{C}^{\bullet}(\mathcal{F}|_P))$ using Euclidean distance

```python
# Note that for 2-parameter case, persistence modules induced by Cech and delaunay are isomorphic.
def to_cech(s,degree):
    _b = s.get_boundaries()
    _d = s.get_dimensions()
    _f = np.asarray(s.get_filtrations())
    _f[:,0] *= 2
    return type(s)(_b,_d,_f)
    
s2 = mp.filtrations.DelaunayLowerstar(x, codensity, reduce_degree=1)
function_cech = to_cech(s2,1)
barcodes_delaunay = s2.persistence_on_lines(
    basepoints=basepoints, 
    directions=directions
)
barcodes_delaunay = tuple(np.array(b[1]).clip(None,height_threshold) for b in barcodes_delaunay)
errors_delaunay = [gd.bottleneck_distance(np.array(b), ground_truth) for b in barcodes_delaunay]

plt.xlabel(fr"Scale $\delta$")
plt.ylabel("Height")
plt.hlines(ground_truth[:,0], xmin=-2,xmax=threshold_radius, linestyles="--", color="k", lw=2)
plt.legend()
mp.module_approximation(s2).plot(1, box=box)

plt.savefig(f"./output/2circles/2circles_{"" if noise else "no"}noise_delaunay_error.pdf")
```

## $H_1(\mathcal{R}^{\bullet\to 2\bullet}(\mathcal{F}|_P))$ using geodesic distance

```python
def compute_geodesic_distance(center, radius, points):
    """
    Compute the geodesic (arc-length) distance matrix between points on a circle.

    Args:
        center (tuple or array-like): (cx, cy) coordinates of the circle center.
        radius (float): Circle radius (> 0).
        points (array-like): Shape (N, 2). Each row is a point (x, y).

    Returns:
        np.ndarray: (N, N) geodesic distance matrix along the circle (arc length).
                    Diagonal is 0.0.
    Notes:
        - Points are *radially projected* to the circle to be robust to noise.
        - Distance between points at angles Î¸i, Î¸j is r * min(|Î”Î¸|, 2Ï€ - |Î”Î¸|).
    """
    points = np.asarray(points, dtype=float)
    c = np.asarray(center, dtype=float).reshape(1, 2)
    r = float(radius)
    if r <= 0:
        raise ValueError("radius must be positive")
    if points.ndim != 2 or points.shape[1] != 2:
        raise ValueError("points must have shape (N, 2)")

    # Vectors from center and radial projection to the circle
    V = points - c
    norms = np.linalg.norm(V, axis=1, keepdims=True)
    # avoid division by zero for points exactly at the center
    V = V / np.maximum(norms, 1e-15) * r

    # Angles in [-pi, pi]
    theta = np.arctan2(V[:, 1], V[:, 0])  # (N,)

    # Pairwise wrapped angle differences in [0, pi]
    dtheta = np.abs(theta[:, None] - theta[None, :])
    dtheta = np.minimum(dtheta, 2 * np.pi - dtheta)

    # Arc-length distances
    D = r * dtheta
    np.fill_diagonal(D, 0.0)
    return D
```

```python
threshold_distance = 100

x1 = x[:npts_per_circle[0],:]
x2 = x[npts_per_circle[0]:,:]
D1 = compute_geodesic_distance(centers[0], radii[0], x1)
D2 = compute_geodesic_distance(centers[1], radii[1], x2)


n1, n2 = D1.shape[0], D2.shape[0]
D = np.full((n1 + n2, n1 + n2), threshold_distance, dtype=float)

# place D1 and D2 along the diagonal
D[:n1, :n1] = D1
D[n1:, n1:] = D2

# set diagonals to 0 (optional safety)
np.fill_diagonal(D, 0.0)

degree=1
st = mp.filtrations.RipsLowerstar(distance_matrix=D, function=x[:,1], threshold_radius=(2/2)*np.pi*threshold_radius)
st.collapse_edges(-1)
st.expansion(degree+1)
s_imgrips = mp.Slicer(st).minpres(degree)
s_imgrips = to_rbga(s_imgrips, degree)
s_imgrips = fix_image(s_imgrips, degree).minpres(degree)
```

```python
barcodes_imrips_geodesic = s_imgrips.persistence_on_lines(
    basepoints=basepoints, 
    directions=directions)

barcodes_imrips_geodesic = tuple(np.array(b[1]).clip(None,height_threshold) for b in barcodes_imrips_geodesic)
errors_imrips_geodesic = [gd.bottleneck_distance(np.array(b), ground_truth) for b in barcodes_imrips_geodesic]

mp.module_approximation(s_imgrips).plot(1, box=box)
plt.xlabel(fr"Scale $\delta$")
plt.ylabel(r"Height")
plt.hlines(ground_truth[:,0], xmin=-2,xmax=threshold_radius, linestyles="--", color="k", lw=2)
plt.legend()

plt.savefig(f"./output/2circles/2circles_{"" if noise else "no"}noise_geodesic_rips_image_error.pdf")
```
